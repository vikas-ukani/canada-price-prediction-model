```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline

from sklearn import linear_model
```

```python
df = pd.read_csv('canada_per_capita_income.csv', delimiter=',')
df
```

```python
plt.xlabel("Years ")
plt.ylabel("Income")
plt.title("Year wise Income In CANADA")
plt.scatter(df['year'], df['per capita income (US$)'])
plt.show()
```

```python
new_df = df.drop(columns='per capita income (US$)', inplace=False)
new_df
```

```python
# Creating Instance of Linear Regression Model
model = linear_model.LinearRegression()
```

```python
# Fit The Data As per the requirments 
model = model.fit(new_df, df['per capita income (US$)'])
```

```python
# Finnaly Prediction The Values 
model.predict([[2020]])
```
